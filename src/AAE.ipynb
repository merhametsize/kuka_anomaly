# Model definition
LSTM and CNN models

```python
input_shape = (data_normal.shape[1], 1) if mode == 'mean' else (data_normal.shape[1], data_normal.shape[2])
if mode == 'mean':
    hid_dim1 = int(data_normal.shape[1]/2)
    hid_dim2 = int(data_normal.shape[1]/5)
else:
    hid_dim1 = int(data_normal.shape[2]/2)
    hid_dim2 = int(data_normal.shape[2]/5)
print(f'Input shape: {input_shape}')
print(f'Hidden dim 1: {hid_dim1}')
print(f'Hidden dim 2: {hid_dim2}')

if model == 'lstm':
    encoder = keras.models.Sequential(name='encoder')
    encoder.add(layers.LSTM(hid_dim1, activation='tanh', return_sequences=True, input_shape=input_shape))
    encoder.add(layers.LSTM(hid_dim2, activation='tanh', return_sequences=False, input_shape=input_shape))

    decoder = keras.models.Sequential(name='decoder')
    decoder.add(layers.RepeatVector(sequence_length))
    decoder.add(layers.LSTM(hid_dim2, activation='tanh', return_sequences=True))
    decoder.add(layers.LSTM(hid_dim1, activation='tanh', return_sequences=True))
    decoder.add(layers.TimeDistributed(layers.Dense(input_shape[1])))

    discriminator = keras.models.Sequential(name='discriminator')
    #discriminator.add(layers.LSTM(20, activation='tanh', return_sequences=False))
    discriminator.add(layers.Dense(16, activation='relu'))
    discriminator.add(layers.Dropout(0.2))
    discriminator.add(layers.Dense(1, activation='sigmoid'))
elif model == 'conv':
    encoder = keras.models.Sequential(name='encoder')
    encoder.add(layers.Conv1D(filters=hid_dim1, kernel_size=3, padding="same", strides=2, activation="relu", input_shape=input_shape))
    encoder.add(layers.Conv1D(filters=hid_dim2, kernel_size=3, padding="same", strides=2, activation="relu"))

    decoder = keras.models.Sequential(name='decoder')
    decoder.add(layers.Conv1DTranspose(filters=hid_dim2, kernel_size=3, padding="same", strides=2, activation="relu"))
    decoder.add(layers.Conv1DTranspose(filters=hid_dim1, kernel_size=3, padding="same", strides=2, activation="relu"))
    decoder.add(layers.Conv1DTranspose(filters=data_normal.shape[2], kernel_size=3, padding="same", activation="sigmoid"))

    discriminator = keras.models.Sequential(name='discriminator')
    discriminator.add(layers.Flatten())
    discriminator.add(layers.Dense(16, activation='relu'))
    discriminator.add(layers.Dropout(0.2))
    discriminator.add(layers.Dense(1, activation='sigmoid'))

autoencoder = keras.models.Sequential()
autoencoder.add(encoder)
autoencoder.add(decoder)

encoder_discriminator = keras.models.Sequential()
encoder_discriminator.add(encoder)
encoder_discriminator.add(discriminator)

```

```python
encoder.summary()
decoder.summary()
discriminator.summary()
```

```python
discriminator.compile(optimizer='adam', loss='binary_crossentropy')
autoencoder.compile(optimizer='adam', loss='binary_crossentropy')
encoder_discriminator.compile(optimizer='adam', loss='binary_crossentropy')
```

# Training

```python
batchsize = int(len(data_normal) / 5)

def settrainable(model, toset):
    for layer in model.layers:
        layer.trainable = toset
    model.trainable = toset

for epoch in range(epochs):
    np.random.shuffle(data_normal)
    validation_data = data_normal[np.random.choice(len(data_normal), 1000), :]

    for i in range(int(len(data_normal) / batchsize)):
        settrainable(autoencoder, True)
        settrainable(encoder, True)
        settrainable(decoder, True)

        batch = data_normal[i*batchsize:i*batchsize+batchsize]
        autoencoder.train_on_batch(batch, batch)

        settrainable(discriminator, True)
        batchpred = encoder.predict(batch)
        if model == 'conv':
            fakepred = np.random.standard_normal((batchsize, encoder.output_shape[1], encoder.output_shape[2]))
        else: #model == 'lstm':
            fakepred = np.random.standard_normal((batchsize, encoder.output_shape[1]))
        discbatch_x = np.concatenate([batchpred, fakepred])
        discbatch_y = np.concatenate([np.zeros(batchsize), np.ones(batchsize)])
        discriminator.train_on_batch(discbatch_x, discbatch_y)

        settrainable(encoder_discriminator, True)
        settrainable(encoder, True)
        settrainable(discriminator, False)
        encoder_discriminator.train_on_batch(batch, np.ones(batchsize))

    print (f"{epoch+1}/{epochs} Reconstruction Loss:", autoencoder.evaluate(validation_data, validation_data))
    print (f"{epoch+1}/{epochs} Adverserial Loss:", encoder_discriminator.evaluate(validation_data, np.ones(len(validation_data))))
```

# Anomaly detection
Histogram plots with the reconstruction loss for both normal and anomalous sequences

```python
n_test_samples = 500
```

```python
random_indexes = np.random.choice(len(data_normal), size=n_test_samples)
test_data = data_normal[random_indexes, :]
rec_normal = autoencoder.predict(test_data)

if mode == 'mean':
    normal_loss = np.mean(np.abs(rec_normal - test_data), axis=1)
else:
    normal_loss = np.mean(np.abs(rec_normal - test_data), axis=2)
    normal_loss = np.mean(normal_loss, axis=1)

plt.hist(normal_loss, bins=50)
plt.xlabel("Normal loss")
plt.ylabel("No of examples")
```

```python
random_indexes = np.random.choice(len(data_slow), size=n_test_samples)
test_data = data_slow[random_indexes, :]
rec_slow = autoencoder.predict(test_data)

if mode == 'mean':
    anomalous_loss = np.mean(np.abs(rec_slow - test_data), axis=1)
else:
    anomalous_loss = np.mean(np.abs(rec_slow - test_data), axis=2)
    anomalous_loss = np.mean(anomalous_loss, axis=1)

plt.hist(anomalous_loss, bins=50)
plt.xlabel("Anomalous loss")
plt.ylabel("No of examples")
```

### Threshold fine-tuning

```python
threshold = 0.43

fig, ax = plt.subplots()
ax.scatter(np.arange(n_test_samples), normal_loss, label='Normal')
ax.scatter(np.arange(n_test_samples, n_test_samples*2), anomalous_loss, label='Anomalous')
ax.hlines(threshold, ax.get_xlim()[0], ax.get_xlim()[1], colors="r", zorder=100, label='Threshold')
ax.legend()
plt.title("Reconstruction error")
plt.ylabel("Reconstruction error")
plt.xlabel("Sample index")
```

### Metrics

```python
loss = np.concatenate((normal_loss, anomalous_loss))
target = np.concatenate((np.ones(n_test_samples), np.zeros(n_test_samples)))
pred = np.less(loss, threshold)

print(f'F1: {f1_score(target, pred, average="macro"):.3f}')
print(f'Accuracy: {accuracy_score(target, pred):.3f}')
print(f'Precision: {precision_score(target, pred, average="macro"):.3f}')
print(f'Recall: {recall_score(target, pred, average="macro"):.3f}')
```
